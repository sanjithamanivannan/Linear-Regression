### Correlation matrix of the dataset
A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables.

```python
dataset.corr()
```

The correlation matrix shows a high correlation of '0.984081' between the columns 'fruitset' and 'yield'.

Hence, the 'fruitset' column has been chosen to predict the yield by using Linear Regression

### Dropping other columns

```python
dataset = dataset.drop(['Row#','clonesize','honeybee','bumbles','andrena','osmia','MaxOfUpperTRange','MinOfUpperTRange','AverageOfUpperTRange','MaxOfLowerTRange','MinOfLowerTRange','AverageOfLowerTRange','RainingDays','AverageRainingDays','fruitmass','seeds'],axis = 1)
```

```python
dataset
```

### Heatmap
Visual representation of the correlation between 'fruitset' and 'yield' using heatmap

```python
plt.figure(figsize=(10,5))
c=dataset.corr()
sn.heatmap(c,cmap='BrBG',annot=True)
```

### Slicing the dataset

```python
x = dataset.iloc[:,0]
```

```python
x
```

```python
x.shape
```

```python
x = dataset.iloc[:,0].values.reshape(-1,1)
```

```python
x.shape
```

```python
y = dataset.iloc[:,-1].values.reshape(-1,1)
```

```python
y.shape
```

```python
y
```

### Scatter plot

```python
plt.scatter(x,y)
plt.show()
```

### Divide the dataset into training and test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Performing Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

### Predicting the yield

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

### Checking the prediction

```python
check = pd.DataFrame(x_test,columns = ['Fruitset'])
```

```python
check['Actual Yield'] = y_test
```

```python
check['Predicted Yield'] = y_pred
```

```python
check
```

### Visualizing the regressor line

```python
plt.scatter(x,y,color = 'orange')
plt.plot(x_test,y_pred,color = 'red')
plt.show()
```

```python
plt.scatter(x_test,y_test,color = 'blue')
plt.plot(x_test,y_pred,color = 'red')
plt.show()
```
